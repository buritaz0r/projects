**Rougher process** (–∞–Ω–≥–ª. ¬´–≥—Ä—É–±—ã–π –ø—Ä–æ—Ü–µ—Å—Å¬ª) ‚Äî —Ñ–ª–æ—Ç–∞—Ü–∏—è  
**Rougher tails** ‚Äî –æ—Ç–≤–∞–ª—å–Ω—ã–µ —Ö–≤–æ—Å—Ç—ã  
**Float banks** ‚Äî —Ñ–ª–æ—Ç–∞—Ü–∏–æ–Ω–Ω–∞—è —É—Å—Ç–∞–Ω–æ–≤–∫–∞  
**Cleaner process** ‚Äî –æ—á–∏—Å—Ç–∫–∞  
**Rougher Au** ‚Äî —á–µ—Ä–Ω–æ–≤–æ–π –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç –∑–æ–ª–æ—Ç–∞  
**Final Au** ‚Äî —Ñ–∏–Ω–∞–ª—å–Ω—ã–π –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç –∑–æ–ª–æ—Ç–∞

**feed size** ‚Äî —Ä–∞–∑–º–µ—Ä –≥—Ä–∞–Ω—É–ª —Å—ã—Ä—å—è  
**feed rate** ‚Äî —Å–∫–æ—Ä–æ—Å—Ç—å –ø–æ–¥–∞—á–∏  

**–ù–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤**  

–ù–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Ç–∞–∫–æ–µ: [—ç—Ç–∞–ø].[—Ç–∏–ø_–ø–∞—Ä–∞–º–µ—Ç—Ä–∞].[–Ω–∞–∑–≤–∞–Ω–∏–µ_–ø–∞—Ä–∞–º–µ—Ç—Ä–∞]. –ü—Ä–∏–º–µ—Ä: rougher.input.feed_ag

**rougher** ‚Äî —Ñ–ª–æ—Ç–∞—Ü–∏—è  
**primary_cleaner** ‚Äî –ø–µ—Ä–≤–∏—á–Ω–∞—è –æ—á–∏—Å—Ç–∫–∞

### –ü—Ä–æ–≤–µ—Ä—å—Ç–µ, —á—Ç–æ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –æ–±–æ–≥–∞—â–µ–Ω–∏—è —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–∞ –ø—Ä–∞–≤–∏–ª—å–Ω–æ. –í—ã—á–∏—Å–ª–∏—Ç–µ –µ—ë –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ –¥–ª—è –ø—Ä–∏–∑–Ω–∞–∫–∞ rougher.output.recovery. –ù–∞–π–¥–∏—Ç–µ MAE –º–µ–∂–¥—É –≤–∞—à–∏–º–∏ —Ä–∞—Å—á—ë—Ç–∞–º–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∞. –û–ø–∏—à–∏—Ç–µ –≤—ã–≤–æ–¥—ã.



–≠—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –æ–±–æ–≥–∞—â–µ–Ω–∏—è —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ –≤—ã—à–µ.   

–≥–¥–µ:  
C ‚Äî –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ –ø–æ—Å–ª–µ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏;
F ‚Äî –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ —Å—ã—Ä—å–µ/–∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ –¥–æ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏;  
T ‚Äî –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ –æ—Ç–≤–∞–ª—å–Ω—ã—Ö —Ö–≤–æ—Å—Ç–∞—Ö –ø–æ—Å–ª–µ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏.  
–î–ª—è –ø—Ä–æ–≥–Ω–æ–∑–∞ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏ –¥–æ–ª—é –∑–æ–ª–æ—Ç–∞ –≤ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–∞—Ö –∏ —Ö–≤–æ—Å—Ç–∞—Ö. –ü—Ä–∏—á—ë–º –≤–∞–∂–µ–Ω –Ω–µ —Ç–æ–ª—å–∫–æ —Ñ–∏–Ω–∞–ª—å–Ω—ã–π –ø—Ä–æ–¥—É–∫—Ç, –Ω–æ –∏ —á–µ—Ä–Ω–æ–≤–æ–π –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç.  

````python
# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ recovery —Å–æ–≥–ª–∞—Å–Ω–æ —Ñ–æ—Ä–º—É–ª–µ
C = train_data['rougher.output.concentrate_au']
F = train_data['rougher.input.feed_au']
T = train_data['rougher.output.tail_au']
recovery_calculated = (C * (F - T)) / (F * (C - T)) * 100

# –†–µ–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è recovery
recovery_actual = train_data['rougher.output.recovery']

# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π
mean_recovery_actual = recovery_actual.mean()
mean_recovery_calculated = recovery_calculated.mean()

# –í—ã–≤–æ–¥ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π
print(f'–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ recovery_actual: {mean_recovery_actual:.4f}')
print(f'–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ recovery_calculated: {mean_recovery_calculated:.4f}')
````

````python
# –ù–∞–π–¥—ë–º MAE –º–µ–∂–¥—É –º–æ–∏–º–∏ —Ä–∞—Å—á—ë—Ç–∞–º–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∞ rougher.output.recovery:
# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ MAE
mae = mean_absolute_error(train_data['rougher.output.recovery'], recovery_calculated)

# –í—ã–≤–æ–¥ MAE
print(f'MAE –º–µ–∂–¥—É —Ä–∞—Å—á–µ—Ç–∞–º–∏ –∏ —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ recovery: {mae:}')
````

## –ê–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö

### –ü–æ—Å–º–æ—Ç—Ä–∏—Ç–µ, –∫–∞–∫ –º–µ–Ω—è–µ—Ç—Å—è –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏—è –º–µ—Ç–∞–ª–ª–æ–≤ (Au, Ag, Pb) –Ω–∞ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —ç—Ç–∞–ø–∞—Ö: –≤ —Å—ã—Ä—å–µ, –≤ —á–µ—Ä–Ω–æ–≤–æ–º –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ, –≤ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ –ø–æ—Å–ª–µ –ø–µ—Ä–≤–æ–π –æ—á–∏—Å—Ç–∫–∏ –∏ –≤ —Ñ–∏–Ω–∞–ª—å–Ω–æ–º –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ. –ö–∞–∫–∏–µ –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏ –∏–º–µ—é—Ç —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è? –û–ø–∏—à–∏—Ç–µ –≤—ã–≤–æ–¥—ã.

–î–∞–Ω–Ω—ã–π –∞–Ω–∞–ª–∏–∑ —Ñ–æ–∫—É—Å–∏—Ä—É–µ—Ç—Å—è –Ω–∞ –∏–∑–º–µ–Ω–µ–Ω–∏–∏ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏–∏ —Ç—Ä–µ—Ö –º–µ—Ç–∞–ª–ª–æ–≤ (–∑–æ–ª–æ—Ç–æ - Au, —Å–µ—Ä–µ–±—Ä–æ - Ag, —Å–≤–∏–Ω–µ—Ü - Pb) –Ω–∞ —Ä–∞–∑–Ω—ã—Ö —ç—Ç–∞–ø–∞—Ö –ø—Ä–æ—Ü–µ—Å—Å–∞ –æ–±—Ä–∞–±–æ—Ç–∫–∏ —Ä—É–¥—ã: –≤ —Å—ã—Ä—å–µ, –≤ —á–µ—Ä–Ω–æ–≤–æ–º –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ, –≤ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ –ø–æ—Å–ª–µ –ø–µ—Ä–≤–æ–π –æ—á–∏—Å—Ç–∫–∏ –∏ –≤ —Ñ–∏–Ω–∞–ª—å–Ω–æ–º –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ.

–î–ª—è —ç—Ç–æ–≥–æ –±—ã–ª–∏ –∏–∑–≤–ª–µ—á–µ–Ω—ã —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –º–µ—Ç–∞–ª–ª–∞ –Ω–∞ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —ç—Ç–∞–ø–∞—Ö –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∏–∑ –¥–∞—Ç–∞—Å–µ—Ç–∞ full_data.

````python
# –í—ã–±–∏—Ä–∞–µ–º —Å—Ç–æ–ª–±—Ü—ã —Å –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏–µ–π –∫–∞–∂–¥–æ–≥–æ –º–µ—Ç–∞–ª–ª–∞ –Ω–∞ —Ä–∞–∑–Ω—ã—Ö —ç—Ç–∞–ø–∞—Ö
concentration_au = full_data[['rougher.input.feed_au',
                              'rougher.output.concentrate_au',
                              'primary_cleaner.output.concentrate_au',
                              'final.output.concentrate_au']]

concentration_ag = full_data[['rougher.input.feed_ag',
                              'rougher.output.concentrate_ag',
                              'primary_cleaner.output.concentrate_ag',
                              'final.output.concentrate_ag']]

concentration_pb = full_data[['rougher.input.feed_pb',
                              'rougher.output.concentrate_pb',
                              'primary_cleaner.output.concentrate_pb',
                              'final.output.concentrate_pb']]

# –°–æ–∑–¥–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º —Å —Ä–∞–∑–Ω—ã–º–∏ —Ü–≤–µ—Ç–∞–º–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —ç—Ç–∞–ø–∞
def plot_concentration_distribution(data, element):
    plt.figure(figsize=(12, 6))
    plt.title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏–∏ {element} –Ω–∞ —Ä–∞–∑–Ω—ã—Ö —ç—Ç–∞–ø–∞—Ö –æ–±—Ä–∞–±–æ—Ç–∫–∏')
    colors = ['blue', 'orange', 'green', 'red']
    
    for i, column in enumerate(data.columns):
        stage = ''
        if i == 0:
            stage = '–°—ã—Ä—å–µ'
        elif i == 1:
            stage = '–ß–µ—Ä–Ω–æ–≤–æ–π –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç'
        elif i == 2:
            stage = '–ö–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç –ø–æ—Å–ª–µ –ø–µ—Ä–≤–æ–π –æ—á–∏—Å—Ç–∫–∏'
        elif i == 3:
            stage = '–§–∏–Ω–∞–ª—å–Ω—ã–π –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç'
        sns.histplot(data[column], label=stage, bins=100, kde=True, color=colors[i])
    
    plt.xlabel(f'–ö–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏—è {element}')
    plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
    plt.legend(title='–≠—Ç–∞–ø—ã –æ–±—Ä–∞–±–æ—Ç–∫–∏', loc='upper left')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

# –°—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –º–µ—Ç–∞–ª–ª–∞ —Å —Ä–∞–∑–Ω—ã–º–∏ —Ü–≤–µ—Ç–∞–º–∏
plot_concentration_distribution(concentration_au, '–∑–æ–ª–æ—Ç–∞ (Au)')
plot_concentration_distribution(concentration_ag, '—Å–µ—Ä–µ–±—Ä–∞ (Ag)')
plot_concentration_distribution(concentration_pb, '—Å–≤–∏–Ω—Ü–∞ (Pb)')
````

````python
# –≠—Ç–∞–ø—ã –æ–±—Ä–∞–±–æ—Ç–∫–∏
stages = ['–°—ã—Ä—å–µ', '–ß–µ—Ä–Ω–æ–≤–æ–π –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç', '–ö–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç –ø–æ—Å–ª–µ –ø–µ—Ä–≤–æ–π –æ—á–∏—Å—Ç–∫–∏', '–§–∏–Ω–∞–ª—å–Ω—ã–π –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç']

# –°–æ–∑–¥–∞–µ–º –≥—Ä–∞—Ñ–∏–∫
plt.figure(figsize=(10, 6))

plt.plot(stages, ag, marker='o', label='–°–µ—Ä–µ–±—Ä–æ')
plt.plot(stages, pb, marker='o', label='–°–≤–∏–Ω–µ—Ü')
plt.plot(stages, au, marker='o', label='–ó–æ–ª–æ—Ç–æ')

plt.title('–ò–∑–º–µ–Ω–µ–Ω–∏–µ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏–∏ –º–µ—Ç–∞–ª–ª–æ–≤ –Ω–∞ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —ç—Ç–∞–ø–∞—Ö –æ–±—Ä–∞–±–æ—Ç–∫–∏')
plt.xlabel('–≠—Ç–∞–ø—ã –æ–±—Ä–∞–±–æ—Ç–∫–∏')
plt.ylabel('–°—Ä–µ–¥–Ω—è—è –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏—è')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
````

3. **–ó–æ–ª–æ—Ç–æ (Au):**
   - –ö–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏—è –∑–æ–ª–æ—Ç–∞ (Au) —Ç–∞–∫–∂–µ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è —Å —ç—Ç–∞–ø–∞ –Ω–∞ —ç—Ç–∞–ø, –Ω–æ —ç—Ç–æ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –±–æ–ª–µ–µ –±—ã—Å—Ç—Ä–æ, –æ—Å–æ–±–µ–Ω–Ω–æ –ø–æ—Å–ª–µ —ç—Ç–∞–ø–∞ —Ñ–ª–æ—Ç–∞—Ü–∏–∏. –≠—Ç–æ —Ö–æ—Ä–æ—à–∞—è –Ω–æ–≤–æ—Å—Ç—å, —Ç–∞–∫ –∫–∞–∫ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏—è –∑–æ–ª–æ—Ç–∞ –≤ –∫–æ–Ω–µ—á–Ω–æ–º –ø—Ä–æ–¥—É–∫—Ç–µ (—Ñ–∏–Ω–∞–ª—å–Ω–æ–º –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ) —Ä–∞—Å—Ç–µ—Ç –±–æ–ª–µ–µ –±—ã—Å—Ç—Ä—ã–º–∏ —Ç–µ–º–ø–∞–º–∏. –≠—Ç–æ –ø–æ–∑–∏—Ç–∏–≤–Ω–æ —Å —Ç–æ—á–∫–∏ –∑—Ä–µ–Ω–∏—è –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏, –ø–æ—Å–∫–æ–ª—å–∫—É –∫–æ–Ω–µ—á–Ω—ã–π –ø—Ä–æ–¥—É–∫—Ç, —Å–æ–¥–µ—Ä–∂–∞—â–∏–π –∑–æ–ª–æ—Ç–æ, —è–≤–ª—è–µ—Ç—Å—è —Ü–µ–ª–µ–≤—ã–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–º –ø—Ä–æ—Ü–µ—Å—Å–∞ –æ–±–æ–≥–∞—â–µ–Ω–∏—è.

````python
# –°–æ–∑–¥–∞–µ–º —Å–ø–∏—Å–æ–∫ —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏
columns_to_check = [
    'rougher.input.feed_ag', 'rougher.input.feed_pb', 'rougher.input.feed_sol', 'rougher.input.feed_au',
    'rougher.output.concentrate_ag', 'rougher.output.concentrate_pb', 'rougher.output.concentrate_sol', 'rougher.output.concentrate_au',
    'primary_cleaner.output.concentrate_ag', 'primary_cleaner.output.concentrate_pb', 'primary_cleaner.output.concentrate_sol', 'primary_cleaner.output.concentrate_au',
    'final.output.concentrate_ag', 'final.output.concentrate_pb', 'final.output.concentrate_sol', 'final.output.concentrate_au'
]

# –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ –∫–∞–∂–¥–æ–º—É –¥–∞—Ç–∞—Ñ—Ä–µ–π–º—É –∏ —Ñ–∏–ª—å—Ç—Ä—É–µ–º —Å—Ç—Ä–æ–∫–∏, —É–¥–∞–ª—è—è –∑–Ω–∞—á–µ–Ω–∏—è <= 1
for df in [full_data, train_data, test_data]:
    # –ü—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞–ª–∏—á–∏–µ —Å—Ç–æ–ª–±—Ü–æ–≤ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ –ø–µ—Ä–µ–¥ —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–µ–π
    existing_columns = [col for col in columns_to_check if col in df.columns]
    
    if existing_columns:
        df.drop(df[(df[existing_columns] <= 1).any(axis=1)].index, inplace=True)

# –í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ, –≤ –∫–∞–∂–¥–æ–º –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–æ–≤ –±—É–¥—É—Ç —É–¥–∞–ª–µ–Ω—ã —Å—Ç—Ä–æ–∫–∏, –≥–¥–µ —Ö–æ—Ç—è –±—ã –≤ –æ–¥–Ω–æ–º –∏–∑ —É–∫–∞–∑–∞–Ω–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –∑–Ω–∞—á–µ–Ω–∏–µ <= 1.
````

–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H1): –°—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ä–∞–∑–º–µ—Ä–æ–≤ –≥—Ä–∞–Ω—É–ª –∏—Å—Ö–æ–¥–Ω–æ–≥–æ —Å—ã—Ä—å—è –≤ –æ–±—É—á–∞—é—â–µ–π –∏ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∞—Ö —Ä–∞–∑–ª–∏—á–∞—é—Ç—Å—è.  

````python
# –ü—Ä–æ–≤–µ–¥–µ–º t-—Ç–µ—Å—Ç –¥–ª—è –æ—Ü–µ–Ω–∫–∏ —Å—Ö–æ–∂–µ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π
alpha = 0.05  # –£—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
````

````python
t_stat, p_value = stats.ttest_ind(train_data['rougher.input.feed_size'], test_data['rougher.input.feed_size'], equal_var=False)
````

–§—É–Ω–∫—Ü–∏—è Symmetric Mean Absolute Percentage Error (sMAPE) —è–≤–ª—è–µ—Ç—Å—è –º–µ—Ç—Ä–∏–∫–æ–π –¥–ª—è –æ—Ü–µ–Ω–∫–∏ —Ç–æ—á–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–µ–π –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è, –æ—Å–æ–±–µ–Ω–Ω–æ –≤ —Å–ª—É—á–∞—è—Ö, –∫–æ–≥–¥–∞ –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –∏ —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏—Ö –≤–µ–ª–∏—á–∏–Ω –º–æ–≥—É—Ç –±—ã—Ç—å —Ä–∞–≤–Ω—ã –Ω—É–ª—é –∏–ª–∏ –æ—á–µ–Ω—å –±–ª–∏–∑–∫–∏ –∫ –Ω–µ–º—É.

sMAPE –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –∫–∞–∫ —Å—Ä–µ–¥–Ω–µ–µ –∞–±—Å–æ–ª—é—Ç–Ω—ã—Ö –ø—Ä–æ—Ü–µ–Ω—Ç–Ω—ã—Ö –æ—à–∏–±–æ–∫ –¥–ª—è –∫–∞–∂–¥–æ–π –ø–∞—Ä—ã –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –∏ —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π. –û–Ω–∞ –∏–∑–º–µ—Ä—è–µ—Ç –ø—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –æ—Ç —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏—Ö, —É—á–∏—Ç—ã–≤–∞—è –∏—Ö –∞–±—Å–æ–ª—é—Ç–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è.—ä



````python
# –°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é
def sMAPE(y_target, y_predict):
    smape = np.mean(abs(y_target - y_predict) / ((abs(y_target) + abs(y_predict)) / 2)) * 100
    return smape
````

````python
# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ sMAPE
def final_sMAPE(smape_rougher, smape_final):
    return (0.25*smape_rougher + 0.75*smape_final)
````

````python
# –í—ã–¥–µ–ª—è–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏ —Ü–µ–ª–µ–≤—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –¥–ª—è –≥—Ä—É–±–æ–π –æ—á–∏—Å—Ç–∫–∏ –∏ —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ –ø—Ä–æ–¥—É–∫—Ç–∞
features_train_rougher = train_data.drop(features_to_exclude, axis=1).values
target_train_rougher = train_data['rougher.output.recovery'].values

features_train_final = train_data.drop(features_to_exclude, axis=1).values
target_train_final = train_data['final.output.recovery'].values

# –ü–æ–ª—É—á–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ —Ü–µ–ª–µ–≤—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ –≥—Ä—É–±–æ–π –æ—á–∏—Å—Ç–∫–∏ –∏ —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ –ø—Ä–æ–¥—É–∫—Ç–∞ (–¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –Ω–∞ —Ñ–∏–Ω–∞–ª—å–Ω–æ–º —ç—Ç–∞–ø–µ)
features_test = np.array(test_data.drop(['rougher.output.recovery', 'final.output.recovery'], axis=1))
target_test_rougher = np.array(test_data['rougher.output.recovery'])
target_test_final = np.array(test_data['final.output.recovery'])

# –í—ã–≤–æ–¥–∏–º —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ –º–∞—Å—Å–∏–≤–æ–≤ –¥–ª—è –≥—Ä—É–±–æ–π –æ—á–∏—Å—Ç–∫–∏
print("–ì—Ä—É–±–∞—è –æ—á–∏—Å—Ç–∫–∞:")
print("–ü—Ä–∏–∑–Ω–∞–∫–∏:", features_train_rougher.shape)
print("–¶–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è (–≥—Ä—É–±–∞—è –æ—á–∏—Å—Ç–∫–∞):", target_train_rougher.shape)

# –í—ã–≤–æ–¥–∏–º —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ –º–∞—Å—Å–∏–≤–æ–≤ –¥–ª—è —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ –ø—Ä–æ–¥—É–∫—Ç–∞
print("–§–∏–Ω–∞–ª—å–Ω—ã–π –ø—Ä–æ–¥—É–∫—Ç:")
print("–ü—Ä–∏–∑–Ω–∞–∫–∏:", features_train_final.shape)
print("–¶–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è (—Ñ–∏–Ω–∞–ª—å–Ω—ã–π –ø—Ä–æ–¥—É–∫—Ç):", target_train_final.shape)
````

–≠—Ç–æ —Ñ—É–Ω–∫—Ü–∏—è –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ scikit-learn (sklearn), –∫–æ—Ç–æ—Ä–∞—è –ø–æ–∑–≤–æ–ª—è–µ—Ç —Å–æ–∑–¥–∞–≤–∞—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ –æ—Ü–µ–Ω—â–∏–∫–∏ (scorers). –û—Ü–µ–Ω—â–∏–∫–∏ –∏—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è –¥–ª—è –∏–∑–º–µ—Ä–µ–Ω–∏—è –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–µ–π, –∏ –æ–Ω–∏ –º–æ–≥—É—Ç –±—ã—Ç—å —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º–∏ –º–µ—Ç—Ä–∏–∫–∞–º–∏, —Ç–∞–∫–∏–º–∏ –∫–∞–∫ —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è –æ—à–∏–±–∫–∞ (MSE), —Å—Ä–µ–¥–Ω—è—è –∞–±—Å–æ–ª—é—Ç–Ω–∞—è –æ—à–∏–±–∫–∞ (MAE), –∏–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ –º–µ—Ç—Ä–∏–∫–∏, —Ç–∞–∫–∏–µ –∫–∞–∫ sMAPE (—Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∞–±—Å–æ–ª—é—Ç–Ω–æ–µ –ø—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ).

greater_is_better = False: –≠—Ç–æ—Ç –∞—Ä–≥—É–º–µ–Ω—Ç —É–∫–∞–∑—ã–≤–∞–µ—Ç, —á—Ç–æ –º–µ–Ω—å—à–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è sMAPE –ª—É—á—à–µ. 

````python
score = make_scorer(sMAPE, greater_is_better = False) 
````

````python
# –°–æ–∑–¥–∞–µ–º –º–æ–¥–µ–ª–∏ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º StandardScaler
lasso_model = make_pipeline(StandardScaler(), Lasso(random_state=12345))
rf_model = make_pipeline(StandardScaler(), RandomForestRegressor(random_state=12345))
dtr_model = make_pipeline(StandardScaler(), DecisionTreeRegressor(random_state=12345))
````

````python
# –ó–∞–¥–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è –º–æ–¥–µ–ª–µ–π
lasso_params = {'lasso__alpha': [0.001, 0.01, 0.1, 1.0]}
rf_params = {
    'randomforestregressor__n_estimators': [10,50],  
    'randomforestregressor__max_depth': [None, 5, 10]
}
dtr_params = {
    'decisiontreeregressor__max_depth': [None, 10, 20],
    'decisiontreeregressor__min_samples_split': [2, 5, 10],
    'decisiontreeregressor__min_samples_leaf': [1, 2, 4]
}
````

````python
# –°–æ–∑–¥–∞–µ–º –æ–±—ä–µ–∫—Ç—ã GridSearchCV –¥–ª—è Lasso, RandomForestRegressor –∏ LinearRegression –¥–ª—è –≥—Ä—É–±–æ–π –æ—á–∏—Å—Ç–∫–∏ (rougher)
grid_rougher_lasso = GridSearchCV(lasso_model, lasso_params, cv=5, scoring=score)
grid_rougher_rf = GridSearchCV(rf_model, rf_params, cv=5, scoring=score)
grid_rougher_dtr = GridSearchCV(dtr_model, dtr_params, cv=5, scoring=score)

# –°–æ–∑–¥–∞–µ–º –æ–±—ä–µ–∫—Ç—ã GridSearchCV –¥–ª—è Lasso, RandomForestRegressor –∏ LinearRegression –¥–ª—è —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ –ø—Ä–æ–¥—É–∫—Ç–∞ (final)
grid_final_lasso = GridSearchCV(lasso_model, lasso_params, cv=5, scoring=score)
grid_final_rf = GridSearchCV(rf_model, rf_params, cv=5, scoring=score)
grid_final_dtr = GridSearchCV(dtr_model, dtr_params, cv=5, scoring=score)
````

````python
# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –¥–ª—è —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ –ø—Ä–æ–¥—É–∫—Ç–∞
print("–õ—É—á—à–∏–µ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã –∏ –æ—Ü–µ–Ω–∫–∏ –¥–ª—è —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ –ø—Ä–æ–¥—É–∫—Ç–∞ (final):")
print("Lasso:", grid_final_lasso.best_params_, "SMAPE:", -grid_final_lasso.best_score_)
print("RandomForestRegressor:", grid_final_rf.best_params_, "SMAPE:", -grid_final_rf.best_score_)
print("DecisionTreeRegressor:", grid_final_dtr.best_params_, "SMAPE:", -grid_final_dtr.best_score_)
````

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, Lasso - —ç—Ç–æ –ª—É—á—à–∞—è –º–æ–¥–µ–ª—å –∫–∞–∫ –¥–ª—è –≥—Ä—É–±–æ–π –æ—á–∏—Å—Ç–∫–∏, —Ç–∞–∫ –∏ –¥–ª—è —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ –ø—Ä–æ–¥—É–∫—Ç–∞ –Ω–∞ –æ—Å–Ω–æ–≤–µ –¥–∞–Ω–Ω–æ–π –º–µ—Ç—Ä–∏–∫–∏.

````python
final_smape_lasso = final_sMAPE(-grid_rougher_lasso.best_score_, -grid_final_lasso.best_score_)
print('–ø–æ –º–æ–¥–µ–ª–∏ LASSO –∏—Ç–æ–≥–æ–≤–æ–µ sMAPE =', final_smape_lasso)

final_smape_dtr = final_sMAPE(-grid_rougher_dtr.best_score_, -grid_final_dtr.best_score_)
print('–ø–æ –º–æ–¥–µ–ª–∏ DecisionTreeRegressor –∏—Ç–æ–≥–æ–≤–æ–µ sMAPE =', final_smape_dtr)

final_smape_rf = final_sMAPE(-grid_rougher_rf.best_score_, -grid_final_rf.best_score_)
print('–ø–æ –º–æ–¥–µ–ª–∏ RandomForestRegressor –∏—Ç–æ–≥–æ–≤–æ–µ sMAPE =', final_smape_rf)
````

### –ü—Ä–æ–≤–µ—Ä–∏–º –º–æ–¥–µ–ª—å –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ

````python
# –ü–æ–ª—É—á–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –º–æ–¥–µ–ª–∏ Lasso –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ –≥—Ä—É–±–æ–π –æ—á–∏—Å—Ç–∫–∏
predictions_rougher = grid_rougher_lasso.predict(features_test)

# –ü–æ–ª—É—á–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –º–æ–¥–µ–ª–∏ Lasso –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ –ø—Ä–æ–¥—É–∫—Ç–∞
predictions_final = grid_final_lasso.predict(features_test)

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ sMAPE –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ –≥—Ä—É–±–æ–π –æ—á–∏—Å—Ç–∫–∏
smape_rougher = sMAPE(target_test_rougher, predictions_rougher)

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ sMAPE –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ –ø—Ä–æ–¥—É–∫—Ç–∞
smape_final = sMAPE(target_test_final, predictions_final)

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∏—Ç–æ–≥–æ–≤–æ–≥–æ sMAPE –Ω–∞ –æ—Å–Ω–æ–≤–µ sMAPE –¥–ª—è –≥—Ä—É–±–æ–π –æ—á–∏—Å—Ç–∫–∏ –∏ —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ –ø—Ä–æ–¥—É–∫—Ç–∞
final_smape = final_sMAPE(smape_rougher, smape_final)

# –í—ã–≤–æ–¥ –∏—Ç–æ–≥–æ–≤–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è sMAPE
print('sMAPE –º–æ–¥–µ–ª–∏ LASSO –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ =', final_smape)
````

### –ü—Ä–æ–≤–µ—Ä–∏–º –º–æ–¥–µ–ª—å –Ω–∞ –∞–¥–µ–∫–≤–∞—Ç–Ω–æ—Å—Ç—å.

````python
# –°–æ–∑–¥–∞–µ–º –∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω—É—é –º–æ–¥–µ–ª—å, –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞—é—â—É—é —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ –¥–ª—è –≥—Ä—É–±–æ–π –æ—á–∏—Å—Ç–∫–∏
dummy_rougher = DummyRegressor(strategy='mean')
dummy_rougher.fit(features_train_rougher, target_train_rougher)

# –°–æ–∑–¥–∞–µ–º –∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω—É—é –º–æ–¥–µ–ª—å, –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞—é—â—É—é —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ –¥–ª—è —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ –ø—Ä–æ–¥—É–∫—Ç–∞
dummy_final = DummyRegressor(strategy='mean')
dummy_final.fit(features_train_final, target_train_final)

# –ü—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ–º —Å –∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω—ã–º–∏ –º–æ–¥–µ–ª—è–º–∏
dummy_predictions_rougher = dummy_rougher.predict(features_test)
dummy_predictions_final = dummy_final.predict(features_test)

# –í—ã—á–∏—Å–ª—è–µ–º sMAPE –¥–ª—è –∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π
dummy_smape_rougher = sMAPE(target_test_rougher, dummy_predictions_rougher)
dummy_smape_final = sMAPE(target_test_final, dummy_predictions_final)

# –í—ã—á–∏—Å–ª—è–µ–º final_sMAPE –¥–ª—è –∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω–æ–π –º–æ–¥–µ–ª–∏
dummy_final_smape_test = final_sMAPE(dummy_smape_rougher, dummy_smape_final)

print('sMAPE –¥–ª—è –∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω–æ–π –º–æ–¥–µ–ª–∏:', dummy_final_smape_test)
````
